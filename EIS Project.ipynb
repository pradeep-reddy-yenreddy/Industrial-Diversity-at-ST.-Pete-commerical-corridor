```python
import pandas as pd
```

```python
df = pd.read_excel('/Users/pradeepreddy/Downloads/Business Tax ID Data 2015.xlsx')
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
df = df[df['BUSINESS STATUS'] == 'A']
print(df.shape)
```

```python
duplicates = df.duplicated(subset=['ADDRESS1','CONTROL #', 'Industry','PIN #'], keep=False)
```

```python
duplicate_rows = df[duplicates]
print(duplicate_rows)
```

```python
df_unique = df.drop_duplicates(subset=['ADDRESS1', 'Industry','PIN #'], keep='first')
print(df_unique.shape)
```

```python
print(df_unique.columns)
```

```python
_16th_St_Corridor = df_unique[df_unique['16th St Corridor'] == 'Within 16th St Corridor']
_18Th_Ave_S_Corridor = df_unique[df_unique['18Th Ave S Corridor'] == 'Within 18th Ave Corridor']
_22nd_St_S_Corridor = df_unique[df_unique['22nd St S Corridor'] == 'Within 22nd Street South Corridor']
_34th_St_Corridor = df_unique[df_unique['34th St Corridor'] == 'Within 34th St Corridor']
_49th_St_Corridor = df_unique[df_unique['49th St Corridor'] == 'Within 49th St Corridor']
_Dr_MLK_Jr_St_Corridor = df_unique[df_unique['Dr MLK Jr St  Corridor'] == 'Within DrMLK JR ST  Corridor']
_5th_Ave_Corridor = df_unique[df_unique['5th Ave  Corridor'] == 'Within 5th Ave S Corridor']
_Central_Ave_Corridor = df_unique[df_unique['Central Ave Corridor'] == 'Within Central Ave Corridor']
```

```python
_16th_St_Corridor = _16th_St_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_18Th_Ave_S_Corridor = _18Th_Ave_S_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_22nd_St_S_Corridor = _22nd_St_S_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_34th_St_Corridor = _34th_St_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)

_49th_St_Corridor = _49th_St_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_Dr_MLK_Jr_St_Corridor = _Dr_MLK_Jr_St_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_5th_Ave_Corridor = _5th_Ave_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
_Central_Ave_Corridor = _Central_Ave_Corridor.drop(['16th St Corridor', '18Th Ave S Corridor',
       '22nd St S Corridor', '34th St Corridor', '49th St Corridor',
       'Dr MLK Jr St  Corridor', '5th Ave  Corridor', 'Central Ave Corridor'], axis=1)
```
